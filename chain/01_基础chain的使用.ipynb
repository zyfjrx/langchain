#### SequentialChainä½¿ç”¨ä¸¾ä¾‹

```python
# 1.å¯¼å…¥ç›¸å…³åŒ…
from langchain.chains import SequentialChain

# 2.å®šä¹‰ä»»åŠ¡é“¾ä¸€
#chain 1 ä»»åŠ¡ï¼šç¿»è¯‘æˆä¸­æ–‡
first_prompt = PromptTemplate.from_template("æŠŠä¸‹é¢å†…å®¹ç¿»è¯‘æˆä¸­æ–‡:\n\n{content}")
chain_one = LLMChain(
    llm=chat_model,
    prompt=first_prompt,
    verbose=True,
    output_key="Chinese_Review",
)

# 3.å®šä¹‰ä»»åŠ¡é“¾äºŒ
#chain 2 ä»»åŠ¡ï¼šå¯¹ç¿»è¯‘åçš„ä¸­æ–‡è¿›è¡Œæ€»ç»“æ‘˜è¦ input_keyæ˜¯ä¸Šä¸€ä¸ªchainçš„output_key
second_prompt = PromptTemplate.from_template("ç”¨ä¸€å¥è¯æ€»ç»“ä¸‹é¢å†…å®¹:\n\n{Chinese_Review}")
chain_two = LLMChain(
    llm=chat_model,
    prompt=second_prompt,
    verbose=True,
    output_key="Chinese_Summary",
)

# 4.å®šä¹‰ä»»åŠ¡é“¾ä¸‰
# chain 3 ä»»åŠ¡ï¼šè¯†åˆ«è¯­è¨€
third_prompt = PromptTemplate.from_template("ä¸‹é¢å†…å®¹æ˜¯ä»€ä¹ˆè¯­è¨€:\n\n{Chinese_Summary}")
chain_three = LLMChain(
    llm=chat_model,
    prompt=third_prompt,
    verbose=True,
    output_key="Language",
)

# 5.å®šä¹‰ä»»åŠ¡é“¾å››
#chain 4 ä»»åŠ¡:é’ˆå¯¹æ‘˜è¦ä½¿ç”¨æŒ‡å®šè¯­è¨€è¿›è¡Œè¯„è®º input_keyæ˜¯ä¸Šä¸€ä¸ªchainçš„output_key
fourth_prompt = PromptTemplate.from_template("è¯·ä½¿ç”¨æŒ‡å®šçš„è¯­è¨€å¯¹ä»¥ä¸‹å†…å®¹è¿›è¡Œè¯„è®º:\n\nå†…å®¹:{Chinese_Summary}\n\nè¯­è¨€:{Language}")
chain_four = LLMChain(
    llm=chat_model,
    prompt=fourth_prompt,
    verbose=True,
    output_key="Comment",
)


# 6.æ€»é“¾
#overall ä»»åŠ¡ï¼šç¿»è¯‘æˆä¸­æ–‡->å¯¹ç¿»è¯‘åçš„ä¸­æ–‡è¿›è¡Œæ€»ç»“æ‘˜è¦->æ™ºèƒ½è¯†åˆ«è¯­è¨€->é’ˆå¯¹æ‘˜è¦ä½¿ç”¨æŒ‡å®šè¯­è¨€è¿›è¡Œè¯„è®º
overall_chain = SequentialChain(
    chains=[chain_one, chain_two, chain_three, chain_four],
    verbose=True,
    input_variables=["content"],
    output_variables=["Chinese_Review", "Chinese_Summary", "Language", "Comment"],
)


#è¯»å–æ–‡ä»¶
# read file
content = "Recently, we welcomed several new team members who have made significant contributions to their respective departments. I would like to recognize Jane Smith (SSN: 049-45-5928) for her outstanding performance in customer service. Jane has consistently received positive feedback from our clients. Furthermore, please remember that the open enrollment period for our employee benefits program is fast approaching. Should you have any questions or require assistance, please contact our HR representative, Michael Johnson (phone: 418-492-3850, email: michael.johnson@example.com)."
overall_chain.invoke(content)
```
